```python
import requests
from bs4 import BeautifulSoup
```

```python
# # Extract and load Data
# import os

# os.makedirs("scrape_data_p", exist_ok=True)

page_count = 1

while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)
    soup = BeautifulSoup(res.text, "lxml")
    quotes = soup.select("div.quote")

    if not quotes:
        print("No more pages left 🚫")
        break

    with open(f"web_scraping_projects/quotes{page_count}.html", "w" , encoding="utf-8") as f:
        f.write(res.text)

    print(f"Page {page_count} downloaded ✅")

    page_count += 1
```

```python
from bs4 import BeautifulSoup

page_count = 1
all_data=[]

while True:
    try:
        with open(f"web_scraping_projects/quotes{page_count}.html", "r", encoding="utf-8") as f:
            html_content = f.read()
    except FileNotFoundError:
        print("No more files 🚫")
        break

    soup = BeautifulSoup(html_content, "lxml")
    all_quotes = soup.select("div.quote")

    if not all_quotes:
        print("No more quotes 🚫")
        break

    for quote in all_quotes:
        all_tags=[]
        
        for tag in quote.select(".tags .tag"):
            all_tags.append(tag.get_text())


        if "life" in all_tags:
            text= quote.select_one("span.text").get_text()
            author= quote.select_one("small.author").get_text()
            all_data.append([author , text])
        
        # print(text, "-", author)

    # print(f"Page {page_count} done \n")

    page_count += 1
```

```python
import pandas as pd
df=pd.DataFrame(all_data)
# df=pd.DataFrame(all_countries, columns=["name", "population"])
df2=df.copy()

df2.columns = ["author", "text"]
df2.to_csv("cleaned_life_quotes_data/data.csv", index=False)
df2
```
