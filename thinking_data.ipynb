```python
# Load and print data
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data
```

```python
data=clean_data(data)
print(data)

```

# Get meanningful insights from data

```python
def get_insight(data):
    top_rat=0
    for user in data:
        top_rat+=float(user["rating"])
    print(f"Avg rating = {top_rat/len(data)}")
    # fetch poor rating
    poor_rat=0
    for user in data:
        if float(user["rating"]) < 3:
            poor_rat+=1
    print(f"% of poor rating is {poor_rat/len(data) * 100}%")


    strong_rat=0
    for user in data:
        if float(user["rating"]) > 3:
            strong_rat+=1
    print(f"% of poor rating is {strong_rat/len(data) * 100}%")
        
```

```python
get_insight(data)
```

# Build recommendation Feature

```python
def get_recommendations(data):
    recommendation=[]
    for user in data:
        curr_recommen={}
        curr_recommen["name"]=user["name"]
        if float(user["rating"]) > 4:
            curr_recommen["Brand"]="Apple"
        else:
            curr_recommen["Brand"]="Samsung"
        recommendation.append(curr_recommen)
    return recommendation
```

```python
get_recommendations(data)
```
