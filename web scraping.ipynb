```python
from bs4 import BeautifulSoup
```

```python
with open ("Scrape_Data/data.html", "r") as f:
    html_content=f.read()
soup=BeautifulSoup(html_content, "lxml")
```

```python
soup.find('h1')
```

```python
all_h3=soup.find_all('h3')
all_countries=[]
for h3 in all_h3:
    name=h3.get_text(strip=True)
    # print(h3.find_parent('div')['class'])
    population=h3.find_next('div').select('span.country-population')[0].get_text(strip=True)
    # print(h3.get('class'))
    # print(h3.attrs['class'])
    all_countries.append([name,population])
```

```python
all_countries
```

```python
import pandas as pd
df=pd.DataFrame(all_countries)
# df=pd.DataFrame(all_countries, columns=["name", "population"])
df2=df.copy()

df2.columns = ["name", "population"]
df2.to_csv("cleaned_data/data.csv", index=False)
```
