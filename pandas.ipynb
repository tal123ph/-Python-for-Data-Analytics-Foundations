# Merging and joining data

```python
df_customer=pd.DataFrame({
    "customer_id" : [1,2,3,4,5],
    "customer_name" : ["Talha","Umer","Ahsan","M","T"]
})


df_order_id=pd.DataFrame({
    "order_id" : [1,2,3,4,5],
    "customer_id" : [2,4,5,3,1],
    "amount" : [1200,1300,1400,1500,1600]
})

pd.merge(df_customer,df_order_id,on="customer_id")   #inner joining
pd.merge(df_customer,df_order_id,on="customer_id",how="left")   #left joining
pd.merge(df_customer,df_order_id,on="customer_id",how="right")   #right joining
pd.merge(df_customer,df_order_id,on="customer_id",how="outer")   #outer joining
```

# Concatenation in pandas

```python
df_customer=pd.DataFrame({
    "customer_id" : [1,2,3,4,5],
    "customer_name" : ["Talha","Umer","Ahsan","M","T"]
})


df_order_id=pd.DataFrame({
    "order_id" : [1,2,3,4,5],
    "customer_id" : [2,4,5,3,1],
    "amount" : [1200,1300,1400,1500,1600]
})
pd.concat([df_customer,df_order_id]    # row wise
pd.concat([df_customer,df_order_id],axis=1)   #col wise concat
```
